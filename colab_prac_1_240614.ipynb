```python
import gdown
gdown.download('https://bit.ly/3eecMKZ',
               '남산도서관 장서 대출목록 (2021년 04월).csv', quiet=False)
# quiet=False 다운로드 과정 진행 중 진행상황에 대해 아래에 표시해주는 기능
# quiet=True로 사용할 시 표시 안 함
```

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv') as f:
  print(f.readline())

# 파이썬의 open 함수는 기본적으로 텍스트가 'UTF-8'로 저장되어있다고 가정함
# 하지만 한글 텍스트는 EUC-KR을 사용하는 일이 잦다.
# 여기서 발생하는 UnicodeDecodeError다.
```

```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode='rb') as f:
  d = f.readline()
print(chardet.detect(d))

# chardet - 텍스트 인코딩을 감지하는 데 사용
# mode='rb' - 바이너리 모드, 내용을 텍스트가 아닌 바이트 단위로 읽어낸다.
# 인코딩을 감지하기 위해서다.
# readline()을 통해 한 줄을 읽어오고
# chardet.detect를 통해 감지해낸다.
# 결과는 신뢰구간 99%에서 encoding은 EUC-KR로 처리된 한글임을 보여준다.
```

```python
with open('남산도서관 장서 대출목록 (2021년 04월).csv', encoding = 'EUC-KR') as f:
  print(f.readline())
  print(f.readline())

# 레코드(행) 한 줄씩 출력된다.
# 파일이 csv 형태임을 알 수 있다.
```

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding = 'EUC-KR')

'''
아래 오류가 발생하는 이유
pandas는 csv 파일을 읽을 때 같은 열에 어떤 종류의 데이터가 저장되어 있는 지 자동으로 파악한다.
이 때 메모리를 효율적으로 사용하기 위해 csv 파일을 조금씩 나누어 읽는다.
이 과정에서 자동으로 파악한 데이터 타입이 달라지면 경고가 발생한다.
해결 방법은 low_memory=False를 사용해 파일을 한 번에 읽어내는 것이다.
'''
```

```python
import pandas as pd
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding = 'EUC-KR', low_memory=False)
df.head()
# head() 메서드는 데이터를 제대로 불러왔는 지 확인하기 위해 처음 다섯 행을 확인한다.
```
